## Data Split

Since the dataset is time-dependent, we will use time-series train test split. For time-dependent data, random split would lead to data leakage as future information would be used to predict past events, resulting in overly optimistic performance estimates.

```python
from sklearn.model_selection import TimeSeriesSplit
from scipy.stats import ks_2samp
```

```python
def calculate_psi(expected, actual, buckets=10, epsilon=1e-8):
    expected = np.asarray(expected).ravel()
    actual = np.asarray(actual).ravel()
    quantiles = np.linspace(0, 100, buckets + 1)
    breakpoints = np.percentile(expected, quantiles)
    for i in range(1, len(breakpoints)):
        if breakpoints[i] <= breakpoints[i-1]:
            breakpoints[i] = breakpoints[i-1] + 1e-6

    exp_counts, _ = np.histogram(expected, bins=breakpoints)
    act_counts, _ = np.histogram(actual, bins=breakpoints)

    exp_props = (exp_counts + epsilon) / (len(expected) + epsilon * len(exp_counts))
    act_props = (act_counts + epsilon) / (len(actual) + epsilon * len(act_counts))

    psi_vals = (exp_props - act_props) * np.log(exp_props / act_props)
    return float(np.sum(psi_vals))
```

```python
def stability_report(train_df, test_df, numeric_only=True, buckets=10):
    results = []
    cols = train_df.columns
    if numeric_only:
        cols = [c for c in cols if pd.api.types.is_numeric_dtype(train_df[c])]

    for col in cols:
        t_col = train_df[col].dropna().values
        s_col = test_df[col].dropna().values
        if len(t_col) < 10 or len(s_col) < 10:
            continue

        psi = calculate_psi(t_col, s_col, buckets=buckets)
        ks_stat, ks_p = ks_2samp(t_col, s_col)
        results.append({"feature": col, "PSI": psi, "KS_stat": ks_stat, "KS_pvalue": ks_p})
    return pd.DataFrame(results).sort_values("PSI", ascending=False)
```

```python
def drift_over_time(df, features=None, n_splits=5, numeric_only=True, buckets=10, sort_by_idx=True):
    if features is None:
        features = df.columns.tolist()
        if numeric_only:
            features = [c for c in features if pd.api.types.is_numeric_dtype(df[c])]

    X = df[features].reset_index(drop=True)
    tscv = TimeSeriesSplit(n_splits=n_splits)
    psi_list = []
    ks_list = []
    fold_names = []
    for i, (train_idx, test_idx) in enumerate(tscv.split(X)):
        train_df = X.iloc[train_idx]
        test_df = X.iloc[test_idx]
        report = stability_report(train_df, test_df, numeric_only=True, buckets=buckets)
        report = report.set_index("feature")
        psi_list.append(report["PSI"])
        ks_list.append(report["KS_stat"])
        fold_names.append(f"fold_{i+1}")
    psi_df = pd.concat(psi_list, axis=1)
    psi_df.columns = fold_names
    ks_df = pd.concat(ks_list, axis=1)
    ks_df.columns = fold_names
    return psi_df.fillna(0), ks_df.fillna(0)
```

```python
def plot_top_drifts(psi_df, top_n=10):
    mean_psi = psi_df.mean(axis=1).sort_values(ascending=False)
    top_features = mean_psi.head(top_n).index.tolist()
    plt.figure(figsize=(10, 6))
    sns.heatmap(psi_df.loc[top_features], annot=True, fmt=".2f", cmap="Reds")
    plt.title("PSI across folds (top drifting features)")
    plt.xlabel("Fold")
    plt.ylabel("Feature")
    plt.show()

def plot_feature_drift(train_series, test_series, feature_name, bins=50):
    plt.figure(figsize=(8,4))
    sns.histplot(train_series, stat="density", kde=False, bins=bins, label="train", alpha=0.6)
    sns.histplot(test_series,  stat="density", kde=False, bins=bins, label="test",  alpha=0.6)
    plt.legend()
    plt.title(f"Distribution train vs test — {feature_name}")
    plt.show()
```

```python
df_sorted = df.sort_values(by="BILL_AMT1").reset_index(drop=True)

features = [c for c in df_sorted.columns if c not in ["ID", "default.payment.next.month"]]
psi_df, ks_df = drift_over_time(df_sorted[features], features=features, n_splits=5, buckets=10)
psi_mean = psi_df.mean(axis=1).sort_values(ascending=False)
print("Top drifting features by mean PSI:\n", psi_mean)

plot_top_drifts(psi_df, top_n=12)
top_feat = psi_mean.index[0]

X = df_sorted[features].reset_index(drop=True)
tscv = TimeSeriesSplit(n_splits=5)
for train_idx, test_idx in tscv.split(X):
    pass
plot_feature_drift(X.iloc[train_idx][top_feat], X.iloc[test_idx][top_feat], top_feat)
```

## Data Pipeline

I plan to use tree-based models and logistic regression for this problem. Different preprocessing strategies are needed:

**For Tree-Based Models:**
- Robust scaling for numerical features (handles outliers better)
- One-hot encoding for nominal categorical features (SEX, MARRIAGE, EDUCATION)
- Keep ordinal features (PAY_0, PAY_2-PAY_6) as-is since trees can handle ordinal relationships

**For Linear Models:**
- Robust scaling for numerical features
- One-hot encoding for nominal categorical features
- Ordinal encoding for ordinal categorical features
- Log transformation for highly skewed features to reduce impact of outliers

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
# one hot encoder for sex, marriage, education
# no encoder for pay_[i] as tree-based models can handle ordinal categorical features directly

no_order_categorical_cols = ['SEX', 'MARRIAGE', 'EDUCATION']
order_categorical_cols = ['PAY_0'] + [f'PAY_{i}' for i in range(2, 7)]

tree_preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numerical_cols),
        ('onehot', OneHotEncoder(handle_unknown='ignore'), no_order_categorical_cols),
        ('ordinal', 'passthrough', order_categorical_cols)
    ])

tree_pipeline = Pipeline(steps=[('preprocessor', tree_preprocessor)])
```

```python
# one hot encoder for sex, marriage, education
# ordinal encoder for pay_[i] as linear models need numerical input
# log transform for skewed numerical features

linear_preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numerical_cols),
        ('onehot', OneHotEncoder(handle_unknown='ignore'), no_order_categorical_cols),
        ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), order_categorical_cols)
    ])

class LogTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            X[col] = np.log1p(X[col].clip(lower=0))
        return X

skewed_features = [f for f in numerical_cols if abs(X[f].skew()) > 0.75]

linear_pipeline = Pipeline(steps=[('log', LogTransformer(columns=skewed_features)),
                                   ('preprocessor', linear_preprocessor),
                                  ])
```

```python
X_transformed = tree_pipeline.fit_transform(X)

ohe_features = tree_pipeline.named_steps['preprocessor'].named_transformers_['onehot'].get_feature_names_out(no_order_categorical_cols)
ord_features = order_categorical_cols

X_transformed_df = pd.DataFrame(X_transformed, columns=numerical_cols +
                                list(ohe_features) +
                                ord_features)
```

## Model Selection & Training

We'll evaluate multiple algorithms using cross-validation to select the best performing model. F1-score is chosen as the primary metric due to potential class imbalance in credit default prediction.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
tree_models = {
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(eval_metric='logloss'),
    "LightGBM": LGBMClassifier(verbose=-1)
}

linear_models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
}

tree_pipelines = {
    name: Pipeline([
        ('preprocessor', tree_preprocessor),
        ('classifier', model)
    ]) for name, model in tree_models.items()
}

linear_pipelines = {
    name: Pipeline([
        ('log', LogTransformer(columns=skewed_features)),
        ('preprocessor', linear_preprocessor),
        ('classifier', model)
    ]) for name, model in linear_models.items()
}
```

```python
for name, pipe in tree_pipelines.items():
    scores = cross_val_score(pipe, X, y, cv=5, scoring='f1')
    print(f"{name}: mean F1 = {scores.mean():.4f}, std = {scores.std():.4f}")
```

```python
for name, pipe in linear_pipelines.items():
    scores = cross_val_score(pipe, X, y, cv=5, scoring='f1')
    print(f"{name}: mean F1 = {scores.mean():.4f}, std = {scores.std():.4f}")
```

```python
# Summary table of cross-validation results
model_results = []

print("=== Tree-Based Models ===")
for name, pipe in tree_pipelines.items():
    scores = cross_val_score(pipe, X, y, cv=5, scoring='f1')
    model_results.append({
        'Model': name,
        'Type': 'Tree-based',
        'Mean F1': scores.mean(),
        'Std F1': scores.std()
    })

print("\n=== Linear Models ===")
for name, pipe in linear_pipelines.items():
    scores = cross_val_score(pipe, X, y, cv=5, scoring='f1')
    model_results.append({
        'Model': name,
        'Type': 'Linear',
        'Mean F1': scores.mean(),
        'Std F1': scores.std()
    })

# Create results DataFrame
results_df = pd.DataFrame(model_results)
print("\n=== Model Performance Summary ===")
print(results_df.sort_values('Mean F1', ascending=False))
```

```python
tree_best_model = tree_pipelines['LightGBM']
tree_best_model.fit(X_train, y_train)
y_pred = tree_best_model.predict(X_test)
y_proba = tree_best_model.predict_proba(X_test)[:, 1]

linear_best_model = linear_pipelines['Logistic Regression']
linear_best_model.fit(X_train, y_train)
y_pred_lr = linear_best_model.predict(X_test)
y_proba_lr = linear_best_model.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

print("Accuracy (Logistic Regression):", accuracy_score(y_test, y_pred_lr))
print("Precision (Logistic Regression):", precision_score(y_test, y_pred_lr))
print("Recall (Logistic Regression):", recall_score(y_test, y_pred_lr))
print("F1 Score (Logistic Regression):", f1_score(y_test, y_pred_lr))
print("ROC AUC (Logistic Regression):", roc_auc_score(y_test, y_proba_lr))
```

```python
### Model Stability Analysis

# Analyze model performance stability across different time folds
from sklearn.model_selection import TimeSeriesSplit

def analyze_model_stability(X, y, n_splits=5):
    """Analyze model performance across time folds"""
    tscv = TimeSeriesSplit(n_splits=n_splits)
    fold_scores = []
    
    for fold, (train_idx, val_idx) in enumerate(tscv.split(X)):
        X_fold_train, X_fold_val = X.iloc[train_idx], X.iloc[val_idx]
        y_fold_train, y_fold_val = y.iloc[train_idx], y.iloc[val_idx]
        
        # Train pipeline on fold
        pipeline_fold = Pipeline([
            ('preprocessor', tree_preprocessor),
            ('classifier', LGBMClassifier(verbose=-1))
        ])
        pipeline_fold.fit(X_fold_train, y_fold_train)
        
        # Evaluate
        y_pred_fold = pipeline_fold.predict(X_fold_val)
        y_proba_fold = pipeline_fold.predict_proba(X_fold_val)[:, 1]
        
        fold_scores.append({
            'fold': fold + 1,
            'accuracy': accuracy_score(y_fold_val, y_pred_fold),
            'precision': precision_score(y_fold_val, y_pred_fold),
            'recall': recall_score(y_fold_val, y_pred_fold),
            'f1': f1_score(y_fold_val, y_pred_fold),
            'auc': roc_auc_score(y_fold_val, y_proba_fold)
        })
    
    return pd.DataFrame(fold_scores)

# Analyze LightGBM stability
print("=== Model Stability Analysis (LightGBM) ===")
stability_results = analyze_model_stability(X, y)

print(stability_results.round(4))

# Calculate coefficient of variation for each metric
cv_results = {}
for metric in ['accuracy', 'precision', 'recall', 'f1', 'auc']:
    mean_score = stability_results[metric].mean()
    std_score = stability_results[metric].std()
    cv = std_score / mean_score if mean_score != 0 else 0
    cv_results[metric] = cv

print(f"\n=== Coefficient of Variation (lower = more stable) ===")
for metric, cv in cv_results.items():
    stability = "Stable" if cv < 0.1 else "Moderately Stable" if cv < 0.2 else "Unstable"
    print(f"{metric.capitalize()}: {cv:.4f} ({stability})")

# Visualize stability
plt.figure(figsize=(12, 4))
metrics_to_plot = ['f1', 'auc', 'precision', 'recall']
for i, metric in enumerate(metrics_to_plot):
    plt.subplot(1, 4, i+1)
    plt.plot(stability_results['fold'], stability_results[metric], 'o-')
    plt.title(f'{metric.upper()} across folds')
    plt.xlabel('Fold')
    plt.ylabel(metric.upper())
    plt.ylim(0, 1)
    
plt.tight_layout()
plt.show()
```
